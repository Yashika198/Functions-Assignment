1. What is the difference between a function and a method in Python?
->A function is a reusable block of code defined with def (or lambda), independent of any class or instance.
It performs computations or actions using only the explicit parameters passed to it.
def add(a, b):
    return a + b

result = add(3, 5)

A method is essentially a function that is defined inside a class and is bound to objects or the class itself.It may implicitly receive the instance (self) or the class (cls) as its first argument.
class Dog:
    def bark(self):            # instance method
        print("Woof!")

    @classmethod
    def info(cls):            # class method
        print("I'm a Dog class")

2. Explain the concept of function arguments and parameters in Python.
->A parameter (also called a formal parameter) is a variable that appears in the function’s definition. It acts as a placeholder inside the function body for the values.
def add(x, y):      #x and y are parameters
    return x + y

An argument (or actual argument) is the real value or expression passed to the function when it is called. These values are assigned to the corresponding parameters.
add(2, 3)            # 2 and 3 are arguments

3. What are the different ways to define and call a function in python?
-> Ways to Define Functions
a. Named Functions (def)
Standard, reusable functions defined with the def keyword.
def add(a, b):
    return a + b

b. Anonymous (Lambda) Functions
Single-expression functions without a name, defined using lambda.
add = lambda x, y: x + y
print(add(3, 5))

c.Recursive Functions
Functions that call themselves to solve problems by breaking them into smaller subproblems.
def factorial(n):
    return 1 if n == 1 else n * factorial(n - 1)

d. Generator Functions
Defined with def and use yield to lazily produce sequences
def generate_numbers(n):
    for i in range(n):
        yield i

Ways to Call Functions:
a. Positional Arguments
Arguments are passed in order:
def greet(first, last):
    print(first, last)
greet("John", "Doe")

b.Keyword Arguments
Arguments supplied by parameter name; order doesn’t matter:
def greet(first, last):
    print(first, last)
greet(last="Doe", first="John")

c.Default Parameters
Function parameters can have default values:
def greet(name="Guest"):
    print(f"Hello, {name}!")
greet()
greet("Alice")

d.Variable-Length Arguments (*args, **kwargs)
Accept arbitrary numbers of arguments:
def total(*args):
    return sum(args)

def details(**kwargs):
    for k, v in kwargs.items():
        print(k, v)

5. What are iterators in Python and how do they differ from iterables?
->An iterable is any object that can be looped over—such as lists, tuples, dictionaries, sets, strings, and more.
Examples:
[1, 2, 3], "hello", {'a': 1}, range(5), file objects, and generator expressions are all iterables

An iterator is an object that provides access to elements one at a time, maintaining internal state to track the current position.
Examples:
Built-in iterators include the objects returned by iter(my_list), generator objects, and file objects.

Key Differences:
a. Every iterator is an iterable—but not vice versa.
iterators implement both __iter__() and __next__(), while pure iterables may lack __next__()
b. State Maintenance
An iterator tracks iteration state and becomes exhausted once
StopIteration is raised.Iterables don’t track state
c. Re-usability
An iterable like a list can be looped over multiple times.An iterator, once exhausted, cannot be re-used.

6. Explain the concept of generators in Python and how are they defined.
-> A generator is a function that returns an iterator and uses the yield statement to produce values one at a time, pausing between each. Unlike regular functions that use return to give back a value and exit, a generator remembers where it left off and resumes from there on the next call.

Generators can be defined in two main ways:
a. Generator Functions
These are regular functions that use the yield keyword.
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
for number in count_up_to(5):
    print(number)

b. Generator Expressions
These are similar to list comprehensions, but use parentheses () instead of square brackets [].
squares = (x * x for x in range(5))
for square in squares:
    print(square)

7. What are the advantages of using generators over regular functions?
-> a. Memory Efficiency
Generators yield items one at a time, so they don't store the entire output in memory.
b. Lazy Evaluation
Generators compute values only when needed.
This can lead to faster startup and reduced CPU usage if only part of the output is required.
c. Regular functions can't return an infinite list, but generators can yield values forever.
def infinite_counter():
    n = 0
    while True:
        yield n
        n += 1

10. What is the difference between map(), reduce() and filter() functions in Python.
-> a. map() Transforming Items
Purpose: Applies a function to each item in an iterable.
Returns: A map object (iterator) with transformed items.
Use case: Modify or compute a value for every element.
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums)) # Output: [1, 4, 9, 16]

b. filter() Filtering Items
Purpose: Applies a condition (returns True/False) to each item and keeps only those for which the condition is True.
Returns: A filter object (iterator) with filtered items.
Use case: Remove unwanted data based on a condition.
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums)) # Output: [2, 4, 6]

c. reduce()  Reducing to a Single Value
Purpose: Applies a function cumulatively to the items in an iterable, reducing it to a single value.
Returns: A single result.
Must be imported from functools module.
Use case: Perform cumulative operations (e.g., sum, product).
from functools import reduce
nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums) # Output: 24 (1*2*3*4)

```python
#Write a Python function that checks if a given number is prime or not from 1 to 200.
import math

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limit = int(math.isqrt(n)) + 1
    for i in range(3, limit, 2):
        if n % i == 0:
            return False
    return True

def primes_up_to(max_n):
    primes = []
    for num in range(1, max_n + 1):
        if is_prime(num):
            primes.append(num)
    return primes

if __name__ == "__main__":
    primes = primes_up_to(200)
    print("Primes from 1 to 200:")
    print(primes)
```

```python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total terms to generate
        self.count = 0          # Counter for terms generated
        self.a, self.b = 0, 1   # Initial Fibonacci values

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration  # End of iteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=' ')
```

```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent
def gen_exp(exponent):
  for i in range(exponent+1):
    yield 2**i
pow=gen_exp(10)
for num in pow:
  print(num)
```

```python
#Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(file_path):
    with open(file_path, "r") as f:
        for line in f:
            yield line.rstrip("\n")
for line in read_lines("Intro.txt"):
    print(line)
```
